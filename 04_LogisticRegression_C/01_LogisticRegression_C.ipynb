```python
# 모듈 불러오기
from sklearn.linear_model import LogisticRegression

from mglearn.plots import plot_2d_separator
from mglearn.datasets import make_forge

import numpy as np
import matplotlib.pyplot as plt
```

```python
# x, y 데이터 생성하기
x, y = make_forge()

print("x type : {}, shape : {}".format(type(x), x.shape))
print("="*50)

print("[ x ]")
print(x[0])
print("="*50)

print("7 type : {}, shape : {}".format(type(y), y.shape))
print("="*50)

print("[ y ]")
print(y[0])
print("="*50)
```

```python
# LogisticRegression 학습하기
clfs = list()
c_list = [0.001, 1, 1000]
for i in range(3):
    logisticRegression = LogisticRegression(C=c_list[i], max_iter=5000)
    clf = logisticRegression.fit(x, y)
    clfs.append(clf)
```

```python
# 그리기
for i in range(3):
    plt.subplot(1, 3, i+1)
    plot_2d_separator(clfs[i], x, fill=True, eps=0.5, alpha=0.5)


    id_list = list()
    for j in np.unique(y):
        idx = np.where(y==j)[0]
        id_list.append(idx)

    plt.scatter(x[id_list[0]][:, 0], x[id_list[0]][:, 1], marker='^')
    plt.scatter(x[id_list[1]][:, 0], x[id_list[1]][:, 1], marker='o')
```

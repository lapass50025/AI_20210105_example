```python
# 모듈 불러오기
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_breast_cancer
from mglearn.plots import plot_2d_separator

import numpy as np
import matplotlib.pyplot as plt
```

```python
# 데이터 준비하기
cancer = load_breast_cancer()

x_train = cancer.data
y_train = cancer.target

print("[ x_train ]")
print(x_train[0])
print("="*50)

print("[ y_train ]")
print(y_train[0])
```

```python
# 변수 초기화하기
width = cancer.data.shape[1]

line = np.linspace(0, width, num=width).reshape(-1, 1)
```

```python
# LogisticRegression 훈련하기
clfs = list()
logisticRegressions = list()

markers = ['o', '^', 'v']
c_list = [0.01, 1, 100]

for i in range(3):
    log = LogisticRegression(C=c_list[i], penalty='l2', max_iter=5000)
    clf = log.fit(x_train, y_train)

    logisticRegressions.append(log)
    clfs.append(clf)
```

```python
# 그래프 출력하기
for i in range(3):
    plt.subplot(1, 3, i+1)
    plt.scatter(line, clfs[i].coef_, marker=markers[i])
```
